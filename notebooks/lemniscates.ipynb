```python
import numpy as np
import matplotlib.pyplot as plt

def generate_lemniscate(foci, R, N):
    # Функция для вычисления значения P(x, y)
    def lemniscate_equation(x, y):
        result = 1.0
        for focus in foci:
            result *= np.sqrt((x - focus[0])**2 + (y - focus[1])**2)
        return result - R

    # Генерация точек лемнискаты
    x_values = np.linspace(-R, R, N)
    y_values = np.linspace(-R, R, N)
    points = []

    for x in x_values:
        for y in y_values:
            if np.abs(lemniscate_equation(x, y)) < 0.1:  # Простой способ определить, что точка лежит на кривой
                points.append([x, y])

    return np.array(points)
```

```python
focuses=[[2,0],[0,2], [1.8,0.2]]
R=7
N=1000
```

```python
lemniscate_points = generate_lemniscate(focuses, float(R), int(N))

# Визуализация
plt.scatter(lemniscate_points[:, 0], lemniscate_points[:, 1], s=1)
plt.scatter(*zip(*focuses), color='red', marker='o')  # Выводим фокусы
plt.title('Лемниската с фокусами')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.axis('equal')
plt.grid(True)
plt.show()
```
